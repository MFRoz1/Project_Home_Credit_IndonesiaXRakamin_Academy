# Visualization


```python
def plot_visualisasi(x, y):
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(x=x, hue=y, data=data_baru, palette='viridis')

    for p in ax.patches:
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12, color='black')

    # Tambahkan judul dan label
    plt.title(f'Bar Chart: {x} vs {y}', fontsize=16)
    plt.xlabel('Target', fontsize=14)
    plt.ylabel('Count', fontsize=14)

    # Tampilkan legenda
    plt.legend(title='Gender')

    # Tampilkan plot
    plt.show()
```

```python
def plot_histogram(kolom):
    plt.figure(figsize=(10, 6))
    sns.histplot(data_baru[kolom], bins=30, color='skyblue', kde=True, edgecolor='black')  # Ganti 'kolom' dengan nama kolom yang ingin diplot
    plt.title(f'Histogram of {kolom}', fontsize=16)
    plt.xlabel(kolom, fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
plot_visualisasi('TARGET', 'CODE_GENDER')
```

```python
plot_visualisasi('TARGET', 'FLAG_OWN_CAR')
```

```python
plot_histogram('AMT_INCOME_TOTAL')
```
