## Outlier

```python
def detect_outliers(df):
    outliers = pd.DataFrame()
    for kolom in df.columns:
        # Hitung nilai Q1 dan Q3
        Q1 = df[kolom].quantile(0.25)
        Q3 = df[kolom].quantile(0.75)

        # Hitung IQR
        IQR = Q3 - Q1

        # Tentukan batas bawah dan atas untuk mendeteksi outlier
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Filter nilai yang di luar batas
        outliers[kolom] = df[(df[kolom] < lower_bound) | (df[kolom] > upper_bound)][kolom]

    return outliers
```

```python
def outlier(dataframe):
  outlier = dataframe.plot(kind = 'box')
  plt.title("Cek nilai Outlier", size = 17)
  plt.xticks(rotation = 90)
  plt.show()
  return outlier
```

```python
def hapus_outlier(dataframe):
  #Clipping menggunakan Boxplot
  Q1 = dataframe.quantile(0.25)
  Q3 = dataframe.quantile(0.75)
  IQR = Q3 - Q1
  Uwishker = Q3 + 1.5 * IQR
  Lwishker = Q1 + 1.5 * IQR
  clip = dataframe.clip(Lwishker, Uwishker, axis = 1)
  return clip
```
